# Matplotlib tutorial

```python
# check version
import matplotlib
print(matplotlib.__version__)
```

```python
# Most of the matplotlib utilities lies under the `pyplot` submodule
import matplotlib.pyplot as plt
import numpy as np
```

## Pyplot

```python
xpoints = np.array([0, 7, 10])
ypoints = np.array([0, 1250, 1000])
# plotting with line
plt.xlabel("x")
plt.ylabel("y")
plt.plot(xpoints, ypoints)
plt.show()
```

```python
# plotting with dot
plt.plot(xpoints, ypoints, "o")
plt.show()
```

```python
# if we do not specity to points of the x-axis, they will get the default values 0,1,2,...
```

# Markers

```python
plt.plot(xpoints, ypoints, marker="p")
plt.show()
```

```python
# format string `fmt` : `market|line|color`: shortcut string notation parameter to specify the marker
plt.plot(xpoints, ypoints, "p:c", ms = 10, mec="r", mfc="y")
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
```

```python
plt.plot(y1)
plt.plot(y2)
plt.show()
```

# Subplots

```python
# with subplot() you can draw multiple plots in one figure
plt.subplot(2,2,1)
plt.plot(y1)
plt.title("figure 1")

plt.subplot(2,2,2)
plt.plot(y2)
plt.title("figure 2")

plt.subplot(2,2,3)
plt.plot(y2)
plt.title("figure 3")

plt.subplot(2,2,4)
plt.plot(y1)
plt.title("figure 4")

plt.suptitle("FIGURE")
plt.show()
```

# Scatter

```python
# you can use the scatter() to draw a scatter plot.
x = np.random.randint(1, 10, 10)
y = np.random.randint(30, 40, 10)

plt.scatter(x, y, color="red", s=x*100, alpha=0.5)

x = np.random.randint(1, 10, 10)
y = np.random.randint(30, 40, 10)

plt.scatter(x, y)
plt.show()
```

# Bar

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.subplot(1,2, 1)
plt.bar(x,y, color="red", width=0.2)

plt.subplot(1,2,2)
plt.barh(x, y)
plt.show()
```
